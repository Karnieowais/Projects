```python
import numpy
import matplotlib.pyplot as plt

```

```python
x=[1,2,3,4,5]
y1=[0.7606,0.8214,0.6846,0.5987,0.6804]
y2=[0.7524,0.7429,0.7714,0.8095,0.7714]
plt.plot(x,y1, label="Loss")
plt.plot(x,y2, label="Acuracy")
plt.xlabel('Epoch')
plt.ylabel('Loss/Acuraccy')
plt.title('Graph of 150 image training')
plt.legend()
#plt.bar(x,y1)
plt.show()
```
